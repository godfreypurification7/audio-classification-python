```python
import matplotlib.pyplot as plt
%matplotlib inline
import IPython.display as ipd
import librosa 
import librosa.display
```

```python
filename='../dataset/sound/wavs-main/209672-3-8-0.wav'
# D:\audioclassification\dataset
# file_example_WAV_1MG.wav'
```
